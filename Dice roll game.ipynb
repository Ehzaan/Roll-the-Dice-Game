```python
import random
print("lets play a game")
def roll():
    roll = random.randint(1,6)
    return roll

while True:
    no_of_players =input("enter the no of player? (2-4) ")
    if no_of_players.isdigit():
        no_of_players = int(no_of_players)
        if 2<= no_of_players <= 4:
            break
        else:
            print("enter a number btw 2 and 4(inclusive)")
    else:
        print("enter a valid number")
max_score = 50
score_board = [0 for _ in range(no_of_players)]
while max(score_board) < max_score:
    for player_idx in range(0,no_of_players):
        current_score = 0
        print("player {} is good to go!!!".format(player_idx+1))
        while True:
            should_roll = input("do you wanna roll? (y/n)") 
            if should_roll.lower() != "y":
                break
            else: 
                value = roll()
                if value == 1:
                    print("you rolled a 1, turn done")
                    current_score = 0
                    break
                else:
                    current_score = current_score + value
                    print(" you rolled a {}".format(value))
            print("your current score is {}".format(current_score))   
        score_board[player_idx] = score_board[player_idx] + current_score
        print(score_board)
maximum_score = max(score_board)
win_index = score_board.index(maximum_score)
print("player {} has won the game with a score of {}!!!!".format(win_index+1,maximum_score))
    
```
